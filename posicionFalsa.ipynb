```python
import sympy as sp
import matplotlib.pyplot as plt
import numpy as np
import re


print("Indica explicitamente la multiplicación. Ej 4x = (4*x)")
# Solicita al usuario que ingrese una expresión
expresion = input("Ingrese una expresión matemática en términos de X: ").lower()
# Define símbolos
x = sp.symbols('x')

try:
    # Intenta parsear y evaluar la expresión
    resultado = sp.sympify(expresion)
    print(resultado)

    expresion = re.sub(r'(\d+)x', r'\1*x', expresion)

    # Crea una función que puede ser evaluada
    funcion = sp.lambdify(x, resultado, 'numpy')
    print(funcion)
    # Crea un rango de valores para x
    x_vals = np.linspace(-8, 8, 125)  # Puedes ajustar el rango como desees

    # Evalúa la función para los valores de x
    y_vals = funcion(x_vals)

    # Grafica la función
    plt.figure(figsize=(8, 6))
    plt.plot(x_vals, y_vals)

    plt.ylim(-8, 8)

    plt.title("Gráfico de la función")
    plt.xlabel("x")
    plt.ylabel("y")
    plt.grid(True)
    plt.show()
except (sp.SympifyError, ValueError):
    print("La expresión ingresada no es válida.")
```
